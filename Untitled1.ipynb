```python
#import the necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV,train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.svm import SVR
```

```python
train_data=pd.read_csv("/storage/emulated/0/Download/train (3).csv",index_col=0)
test_data=pd.read_csv("/storage/emulated/0/Download/test (2).csv",index_col=0)
```

```python
y=train_data["target"]
y.shape
```

```python
X=train_data.drop(["target"],axis=1)
X.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=0)
```

```python
#models
model1=LinearRegression()
model2=Lasso()
model3=Ridge()
model4=SVR()
param={"alpha":[0.01,0.1,10,100]}
gridlasso=GridSearchCV(model2,param,verbose=1,cv=5)
gridridge=GridSearchCV(model3,param,verbose=1,cv=5)
param_svm={"gamma":[0.01,0.1,1,10,100],"C":[0.01,0.1,1,10,100]}
grid_svm=GridSearchCV(model4,param_svm,verbose=1,cv=5)
```

```python
#linearregression
model1.fit(X_train,y_train)
print("train score "+str(model1.score(X_train,y_train)))
print("test score "+str(model1.score(X_test,y_test)))
```

```python
gridlasso.fit(X_train,y_train)
print("train score "+str(gridlasso.score(X_train,y_train)))
print("test score "+str(gridlasso.score(X_test,y_test)))
```

```python
gridridge.fit(X_train,y_train)
print("test score " +str(gridridge.score(X_test,y_test)))
print("train score "+str(gridridge.score(X_train,y_train)))
```

```python
grid_svm.fit(X_train,y_train)
print (grid_svm.best_params_)
```

```python
print("test score "+str(grid_svm.score(X_test,y_test)))
print("train score " +str(grid_svm.score(X_train,y_train)))
```
