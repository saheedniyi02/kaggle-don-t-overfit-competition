```python
#import the necessary libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.feature_selection import SelectPercentile
```

```python
train_data=pd.read_csv("/storage/emulated/0/Download/train (3).csv",index_col=0)
test_data=pd.read_csv("/storage/emulated/0/Download/test (2).csv",index_col=0)
```

```python
y=train_data["target"]
y.shape
```

```python
X=train_data.drop(["target"],axis=1)
select=SelectPercentile(percentile=50)
select.fit(X,y)
column=X.columns
X=select.transform(X)
selected_index=[]
column_bool=select.get_support()
ind=0
for i in column_bool:
    if i==True:
        selected_index.append(ind)
    ind=ind+1
selected=[]
for i in selected_index:
    selected.append(column[i])
selected
```

```python
test_data=test_data[selected]
test_data.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=0)
```

```python
model1=SVC(random_state=0)
model2=LogisticRegression(random_state=0)
param_lgs={"C":[0.01,0.1,1,10,100]}
param_SVC={"C":[0.01,0.1,1,10,100],"gamma":[0.01,0.1,1,10,100]}
gridsvm=GridSearchCV(model1,param_SVC,verbose=1,cv=5)
gridlgs=GridSearchCV(model2,param_lgs,verbose=1,cv=5)
```

```python
gridsvm.fit(X_train,y_train)
print (gridsvm.best_params_)
```

```python
print("test score "+str(gridsvm.score(X_test,y_test)))
print("train score " +str(gridsvm.score(X_train,y_train)))
```

```python
y_p=gridsvm.predict(X_test) 
```

```python
print(confusion_matrix(y_test,y_p))
#looking at the confusion matrix it only predicted target variables of zero hence SVM might not be the best model
```

```python
#let me try logistic Regression
gridlgs.fit(X_train,y_train)
y_lgs=gridlgs.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_lgs))
print("test score "+str(gridlgs.score(X_test,y_test)))
print("train score "+ str(gridlgs.score(X_train,y_train)))
#it predicted zeros with high accuracy one of zero accuracy
#hence it MIGHT be a good model but we can't be so sure.
#lets try another model is
```

```python
model3=DecisionTreeClassifier(random_state=0)
model4=RandomForestClassifier(random_state=0)
model5=GradientBoostingClassifier(random_state=0)
```

```python
param_tree={"max_depth":[4,5,6,8,10,15,20,40,50,75,100,150],"max_features":[5,6,7,9,10,25,30,50,75,100,125,150]}
param_boosting={"learning_rate":[0.01,0.015,0.05,0.1],"max_depth":[4,5,6,8,10,15,20,40,50,75,100,150],"max_features":[5,6,7,9,10,25,30,50,75,100,125,150]}
param_forest={"max_depth":[4,5,6,8,10,15,20,40,50,75,100,150],"max_features":[5,6,7,9,10,25,30,50,75,100,125,150]}
gridtree=GridSearchCV(model3,param_tree,cv=5,verbose=2)
gridforest=GridSearchCV(model4,param_forest,cv=5,verbose=1)
```

```python
gridtree.fit(X_train,y_train)
y_tree=gridtree.predict(X_test)
```

```python
y_tree=gridtree.predict(X_test)
```

```python
print("test score "+str(gridtree.score(X_test,y_test)))
print("train score "+str(gridtree.score(X_train,y_train)))
print(confusion_matrix(y_test,y_tree))
```

```python
gridforest.fit(X_train,y_train)
```

```python
y_forest=gridforest.predict(X_test)
```

```python
print("test score "+str(gridforest.score(X_test,y_test)))
print("train score "+str(gridforest.score(X_train,y_train)))
print(confusion_matrix(y_test,y_forest))
```
